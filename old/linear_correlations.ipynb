#### Pearson Correlation

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from pylab import rcParams
import seaborn as sb

import scipy
from scipy.stats.stats import pearsonr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

```python
address = '../datasets/staandelamp_realistic.json'
data = pd.read_json(address)
```

```python
data.head()
```

```python
sb.pairplot(data[0:20])
```

```python
name = data['name']
state = data['state']
time = data['time']
```

```python
pearsonr_coefficient, p_value = pearsonr(time, state)
print('Pr %0.3f' % pearsonr_coefficient)
```

```python
corr = data.corr()
corr
```

```python
sb.heatmap(corr)
```

```python
subset = data[0:100]
```

```python
colors = subset['state']

plt.scatter(subset['time'], subset['name'], c=colors, s=10)
plt.colorbar();
```
